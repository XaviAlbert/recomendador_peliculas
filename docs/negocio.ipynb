```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
uno = {
    'Streamberry': 233,
    'Armadillo Prime Video': 205,
    'Mouse+': 183,
    'RKO Max': 80,
    'Pear TV+': 50
}
```

```python
plataformas = list(uno.keys())
suscriptores = list(uno.values())
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_facecolor('black')
sns.barplot(x=plataformas, y=suscriptores, color='red', ax=ax)
plt.xlabel('Plataforma')
plt.ylabel('Usuarios')
plt.title('Usuarios por plataforma de streaming')
ax.grid(color='white', alpha=0.1)
ax.tick_params(axis='x')
plt.show()
```
